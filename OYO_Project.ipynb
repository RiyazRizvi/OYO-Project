```python
# CATEGORICAL - CATEGORICAL ASSOCIATIONS
# Crosstabs
cross_tab_prop = pd.crosstab(index = oyo['availability_30'], columns = oyo['accommodates'], normalize = "index")
cross_tab_prop
```

```python
# Stacked Bar Graphs
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

plt.legend(loc="upper right", ncol=2)
plt.xlabel("cancellation_policy")
plt.ylabel("bed_type")
plt.show()
```

```python
# Stacked Bars with Labels
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

plt.legend(loc="upper right", ncol=2)
plt.xlabel("bed_type")
plt.ylabel("cancellation_policy")

for n, x in enumerate([*cross_tab_prop.index.values]):
    for l in cross_tab_prop.loc[x]:
                
        plt.text(x=n,
                 y=l,
                 s=f'{np.round(l * 100, 1)}%', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")

plt.show()
```

```python
# Heatmaps
col_counts = oyo.groupby(['cancellation_policy', 'bed_type']).size()
col_counts = col_counts.reset_index(name = 'count')
col_counts = col_counts.pivot(index = 'bed_type', columns = 'cancellation_policy', values = 'count')

sns.heatmap(col_counts, annot = True, fmt = 'd')
```
